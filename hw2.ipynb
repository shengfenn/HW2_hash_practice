```python
import matplotlib.pyplot as plt
file = open('hw2_data.txt','r')
c = {}
number = 0


for food in file.readlines():
    food = food.strip()
    if food not in c:
        c[food] = 0
        number += 1
    c[food] += 1
sorted_c = sorted(c.items(), key = lambda x:x[1], reverse = True)
sorted_c_dict = dict(sorted_c)


for i in sorted_c_dict:
    x = i
    y = sorted_c_dict[i]
    plt.bar(x,y)


print("共有",number,"個不重複英文單字")
print("英文單字出現次數如下：",sorted_c_dict)
plt.show()
file.close()
```
