```python
import matplotlib.pyplot as plt
file = open('hw2_data.txt','r')
c = {}
for food in file.readlines():
    food = food.strip()
    if food not in c:
        c[food] = 0
    c[food] += 1
sorted_c = sorted(c.items(), key = lambda x:x[1])
sorted_c_dict = dict(sorted_c)
for i in sorted_c_dict:
    x = i
    y = sorted_c_dict[i]
    plt.bar(x,y)
plt.show()
print(sorted_c_dict)
file.close()
```
